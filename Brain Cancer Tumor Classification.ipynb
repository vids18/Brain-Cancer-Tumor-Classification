# Comparison of different classification algorithms to identify the Brain Cancer Tumor Features

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
X
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# Logistic Regression

## Training Logistic Regression on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier_log = LogisticRegression(random_state = 0)
classifier_log.fit(X_train, y_train)
```

```python
y_pred = classifier_log.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, roc_auc_score
cm_log = confusion_matrix(y_test, y_pred)
print(cm_log)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies_1= cross_val_score(estimator = classifier_log, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies_1.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies_1.std()*100))
```

# Support Vector Machine

## Training Support Vector Machine on the Training set

```python
from sklearn.svm import SVC
classifier_svm = SVC(kernel="rbf", random_state=0, probability = True)
classifier_svm.fit(X_train, y_train)
```

```python
y_pred = classifier_svm.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_svm = confusion_matrix(y_test, y_pred)
print(cm_svm)
accuracy_score(y_test, y_pred)
```

```python
accuracies_1= cross_val_score(estimator = classifier_svm, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies_1.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies_1.std()*100))
```

# Naive Bayes Classification

## Training the Naive Bayes model on Training set

```python
from sklearn.naive_bayes import GaussianNB
classifier_gnb = GaussianNB()
classifier_gnb.fit(X_train, y_train)
```

```python
y_pred = classifier_gnb.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm_gnb = confusion_matrix(y_test, y_pred)
print(cm_gnb)
accuracy_score(y_test, y_pred)
```

```python
accuracies_1= cross_val_score(estimator = classifier_gnb, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies_1.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies_1.std()*100))
```

# Random Forest

## Training the Random Forest model on the training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf= RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier_rf.fit(X_train, y_train)
```

```python
y_pred = classifier_rf.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
cm_gnb = confusion_matrix(y_test, y_pred)
print(cm_gnb)
accuracy_score(y_test, y_pred)
```

```python
accuracies_1= cross_val_score(estimator = classifier_rf, X = X_train, y = y_train, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies_1.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies_1.std()*100))
```

# CatBoost

## Training CatBoost on the Training set

```python
from catboost import CatBoostClassifier
classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
```

```python
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

## Comparing the ROC-AUC values for all the above algorithms

```python
# Getting predicted probabilities
y_score1 = classifier_log.predict_proba(X_test)[:,1]
y_score2 = classifier_svm.predict_proba(X_test)[:,1]
y_score3 = classifier_gnb.predict_proba(X_test)[:,1]
y_score4 = classifier_rf.predict_proba(X_test)[:,1]
y_score5 = classifier.predict_proba(X_test)[:,1]

# Ploting Receiving Operating Characteristic Curve# Creating true and false positive rates
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y_score1, pos_label=2)
false_positive_rate2, true_positive_rate2, threshold2 = roc_curve(y_test, y_score2, pos_label=2)
false_positive_rate3, true_positive_rate3, threshold3 = roc_curve(y_test, y_score3, pos_label=2)
false_positive_rate4, true_positive_rate4, threshold4 = roc_curve(y_test, y_score4, pos_label=2)
false_positive_rate5, true_positive_rate5, threshold5 = roc_curve(y_test, y_score5, pos_label=2)

print('roc_auc_score for Logistic Regression: ', roc_auc_score(y_test, y_score1))
print('roc_auc_score for Support Vector Machine: ', roc_auc_score(y_test, y_score2))
print('roc_auc_score for Naive Bayes: ', roc_auc_score(y_test, y_score3))
print('roc_auc_score for Random Forest: ', roc_auc_score(y_test, y_score4))
print('roc_auc_score for CatBoost: ', roc_auc_score(y_test, y_score5))
```

When comparing the ROC-AUC values, accuracies and standard deviations of all the algorithms I come to a conclusion that Logistic Regression gives the best output for the brain cancer tumor classification. Logistic regression perfects identifies the factors to be benign and malignant in comparison to the other algorithms.
